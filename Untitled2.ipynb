```python
# Se estiver no Colab e quiser enviar o arquivo do Kaggle:
# 1) Baixe 'train.csv' do Titanic no Kaggle
# 2) Envie o arquivo pelo ícone de pasta (ou use files.upload())
# from google.colab import files; files.upload()

import os, io, textwrap
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay, classification_report
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

CSV_CANDIDATES = [
    'train.csv',
    '/content/train - train (6).csv',
    '/kaggle/input/train.csv',
    '/kaggle/input/titanic/train.csv'
]

csv_path = None
for p in CSV_CANDIDATES:
    if os.path.exists(p):
        csv_path = p
        break

if csv_path:
    df = pd.read_csv(csv_path)
    # Tenta converter colunas numéricas para garantir o tipo correto
    # Usa errors='coerce' para transformar valores inválidos em NaN, que o imputer pode tratar.
    for col in ['Age', 'SibSp', 'Parch', 'Fare']:
        if col in df.columns:
            # Remove '.' como separador de milhares e substitui ',' por '.' se for o caso
            # Isso lida com formatos europeus ou inconsistentes antes da conversão para float
            df[col] = df[col].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
            df[col] = pd.to_numeric(df[col], errors='coerce')
    source = f"Arquivo Kaggle detectado: {csv_path}"
else:
    rng = np.random.default_rng(42)
    n = 400
    df = pd.DataFrame({
        'Survived': rng.integers(0, 2, size=n),
        'Pclass': rng.choice([1,2,3], size=n, p=[0.2,0.2,0.6]),
        'Sex': rng.choice(['male','female'], size=n),
        'Age': np.clip(rng.normal(30, 14, size=n), 0, 80),
        'SibSp': rng.integers(0, 5, size=n),
        'Parch': rng.integers(0, 5, size=n),
        'Fare': np.clip(rng.normal(32, 20, size=n), 5, 300),
        'Embarked': rng.choice(['S','C','Q'], size=n, p=[0.72,0.18,0.10])
    })
    source = "Fallback sintético (execute com o CSV do Kaggle para dados reais)"
print(source)
display(df.head())

target = 'Survived'
features = ['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']

X = df[features].copy()
y = df[target].copy()

numeric_features = ['Age','SibSp','Parch','Fare']
numeric_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='median'))])

categorical_features = ['Pclass','Sex','Embarked']
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocess = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

from sklearn import set_config

# Resetar a configuração de saída para o padrão antes de definir e treinar o clf
# Isso evita conflitos se 'transform_output="pandas"' foi definido globalmente antes.
set_config(transform_output="default")

clf = Pipeline(steps=[
    ('prep', preprocess),
    ('model', DecisionTreeClassifier(max_depth=5, random_state=42)) # Alterado max_depth para 5
])

clf.fit(X_train, y_train)
pred = clf.predict(X_test)
acc = accuracy_score(y_test, pred)
print(f"Acurácia (Árvore de Decisão, max_depth=5): {acc:.3f}") # Alterado max_depth na mensagem
ConfusionMatrixDisplay.from_predictions(y_test, pred)
plt.title("Matriz de Confusão — Titanic")
plt.show()

print(classification_report(y_test, pred))
# --- VISUALIZANDO A ÁRVORE (TEXTO) ---  (versão com pandas)
from sklearn import set_config
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# 1) Garantir saída pandas nos transformadores
set_config(transform_output="pandas")

# 2) RECONSTRUIR o preprocess com OneHotEncoder denso (evita erro de sparse no pandas)
numeric_features = ['Age','SibSp','Parch','Fare']
numeric_transformer = Pipeline(steps=[('imputer', SimpleImputer(strategy='median'))])

categorical_features = ['Pclass','Sex','Embarked']
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    # Em scikit-learn >= 1.2 use sparse_output=False; em versões antigas use sparse=False
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocess_pandas = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)

# 3) Transformar X_train em DataFrame já expandido
prep_df = preprocess_pandas.fit_transform(X_train)

# 4) Treinar uma árvore diretamente no espaço transformado para poder imprimir/exportar
tree_for_plot = DecisionTreeClassifier(max_depth=3, random_state=42)
tree_for_plot.fit(prep_df, y_train)

feature_names = list(prep_df.columns)
print(export_text(tree_for_plot, feature_names=feature_names))
# --- PLOT DA ÁRVORE ---
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(14, 7))
plot_tree(
    tree_for_plot,
    feature_names=feature_names,
    class_names=['No','Yes'],
    filled=True,
    rounded=True,
    impurity=True,
    proportion=True
)
plt.title("Estrutura da Árvore (após transformação — saída pandas)")
plt.show()
```
